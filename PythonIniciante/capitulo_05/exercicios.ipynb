<h1 style="color:red">Exercício 5.3</h1>
<p>Faça um programa para escrever a contagem regressiva do lançamento de um foguete. O programa deve imprimir 10, 9, 8,..., 1, 0 e Fogo! na tela. 

```python
x = 10
while x >= 0:
    print(x)
    x = x - 1
    if x == -1:
        print("Fogo!")
```

<h1 style="color:Red">Exercício 5.4</h1>
<p>Modifique o programa anterior para imprimir de 1 até o número digitado pelo usuário, mas dessa vez apenas os números impares.
</p>

```python
fim = int(input("Digite o último número a imprimir: "))
x = 1
while x <= fim:
    if x % 2 == 1:
        print(x)
    x = x + 1
```
