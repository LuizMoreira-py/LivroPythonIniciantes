<h1 style="color:red">Exercício 3.6</h1>
<p>Escreva uma expressão que será utilizada para decidir se um aluno foi ou não aprovado. Para ser aprovado, todas as médias do aluno devem ser maiores que 7. Considere que o aluno cursa apenas três matérias, e que a nota de cada uma está armazenada nas seguintes variáveis: materia1, materia2, materia3.</p>

```python
nome = input("informe o nome do aluno: ")
materia1 = int(input("Informe a nota da materia 1: "))
materia2 = int(input("Informe a nota da materia 2: "))
materia3 = int(input("Informe a nota da materia 3: "))

media = (materia1 + materia2 + materia3) // 3

if media >= 70:
    print("Aluno aprovado")
else:
    print("Aluno reprovado")

print(f"Média das notas: {media}")
```

<h1 style="color:red">Exercício 3.9</h1>
<p>Escreva um programa que leia a quantidade de dias, horas, minutos e segundos do usuário. Calcule o total em segundos.
<p>

```python
dia = int(input("Informe um valor para a variavel dia: "))
hora = int(input("Informe um valor para a variavel hora: "))
minuto = int(input("Informe um valor para a variavel minuto: "))
segundo = int(input("Informe um valor para a variavel segundo: "))

diaSegundo = dia * 86400
horaSegundo = hora * 3600
minutoSegundo = minuto * 60

total = diaSegundo + horaSegundo + minutoSegundo

print(f"O total de segundos em {dia} dias, {hora} horas, {minuto} minutos e {segundo} segundos; é de {total} segundos")
```

<h1 style="color:red">Exercício 3.10</h1>
<p>Faça um programa que calcule o aumento de um salário. Ele deve solicitar o valor do salário e a porcentagem do aumento. Exiba o valor do aumento e do novo salário.
<p>

```python
salario = float(input("Informe um valor para a variavel salário: "))
percentual = float(input("Informe um percentual de aumento: "))

aumento = (salario * percentual) / 100

novoSalario = salario + aumento

print("O aumento do salário é de R$ {:5.2f}". format(aumento))
print("O novo salário é R$ {:5.2f}".format(novoSalario))
```

<h1 style="color:red">Exercício 3.11</h1>
<p>Faça um programa que solicite o preço de uma mercadoria e o percentual de desconto. Exiba o valor do desconto e o preço a pagar.
<p>

```python
precoMercadoria = float(input("Informe um valor para a variavel preço: "))
percentualDesconto = float(input("Informe um percentual de desconto: "))

desconto = (precoMercadoria * percentualDesconto) / 100

valorApagar = precoMercadoria - desconto

print("O valor do desconto é de R$ {:5.2f}".format(desconto))
print("O valor a pagar é de R$ {:5.2f}".format(valorApagar))
```

<h1 style="color:red">Exercício 3.13</h1>
<p>escreva um programa que converta uma temperatura digitada em °C em °F. A fórmula para essa conversõa é:
</p>
<p>F=(9xC/5)+32
</p>

```python
celsius = int(input("Informe um valor para a temperatura em °C: "))

fahrenheit = (9 * celsius // 5) + 32

print(f"{celsius} °C equivale {fahrenheit} °F")
```

<h1 style="color:red">Exercício 3.14</h1>
<p>Escreva um programa que pergunte a quantidade de km percorridos por um carro alugado pelo usuário, assim como a quantidade de dias pelos quais o carro foi alugado. Calcule o preço a pagar, sabendo que o carro custa R$ 60,00 por dia e R$ 0,15 por km rodado.
</p>

```python
aluguelPorDia = 60
aluguelPorKm = 0.15

DiasAluguel = int(input("Informe a quantidade de dias que o carro foi alugado: "))
KMpercorrido = float(input("informe a quantidade de Km percorridos: "))

valorApagar = (aluguelPorDia * DiasAluguel) + (aluguelPorKm * KMpercorrido)

print("O valor a pagar pelo aluguel do carro é de R$ {:5.2f}".format(valorApagar))
```

<h1 style="color:red">Exercício 3.15</h1>
<p>Escreva um programa para calcular a redução do tempo de vida de um fumante. Pergunte a quantidade de cigarros fumados por dia e quantos anos ele já fumou. Considere que um fumante perde 10 minutos de vida a cada cigarro, e calcule quantos dias de vida um fumante perderá. Exiba o total em dias.
</p>

```python
MinutosPerdidos = 10

QuantidadePorDia = int(input("Informe quantos cigarros fumados por dia: "))

AnosFumados = int(input("Informe a quantidade de anos fumados: "))

TotalFumadosAno = QuantidadePorDia * 365

TotalFumadosVida = TotalFumadosAno * AnosFumados

TempoFumado = TotalFumadosVida * MinutosPerdidos

TempoPerdido = TempoFumado // 1440

print(f"O tempo de vida perdido é de {TempoPerdido} dias")
```
