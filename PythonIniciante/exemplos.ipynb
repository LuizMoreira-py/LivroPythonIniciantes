<h1 style="color:lightseagreen">Programa 4.3 - Cálculo do Imposto de Renda</h1>
<p>O Programa 4.3 é bem interessante. Tente executá-lo algumas vezes e compare o valor impresso com o valor calculado por você. Rastreie o programa e tente entender o que ele faz antes de ler o parágrafo seguinte. Verifique o que acontece para salários de R$ 500,00, R$ 1.000,00 e R$ 1.500,00.
</p>

```python
salario = float(input("Digite o salário para o calculo do imposto: "))

base = salario
imposto = 0

if base > 3000:
    imposto = imposto + ((base - 3000) * 0.35)
    base = 3000

if base > 1000:
    imposto = imposto + ((base - 1000) * 0.20)

print(f"Salário: R${salario:6.2f} Imposto a pagar: R${imposto:6.2f}")
```

<h1 style='color:LightSeaGreen'>Programa 4.5 - Conta de telefone com três faixas de preço</h1>
<p>Vejamos o exemplo de calcular a conta de um telefone celular da empresa Tchau.</p>
<p>Os planos da empresa Tchau são bem interessantes e oferecem preços diferenciados de acordo com a quantidade de minutos usados por mês. Abaixo de 200 minutos, a empresa cobra R$ 0,20 por minutos. Entre 200 e 400 minutos, o preço é de R$ 0,18. Acima de 400 minutos, o preço por minuto é de R$ 0,15.</p>

<p>O Programa 4.5 resolve esse problema:</p>

```python
minutos = int(input("Quantos minutos você utilizou este mês: "))

if minutos < 200:
    preco = 0.20

else:
    if minutos < 400:
        preco = 0.18
    else:
        preco = 0.15

print(f"Você vai pagar este mês: R${minutos * preco:6.2f}")
 
```

<h1 style='color:LightSeaGreen'>Programa 5.3 - Múltipla escolha</h1>

<h1 style="color:LightSeaGreen">Programa 5.4 - Contagem de cédulas</h1>
<p>Vejamos como exemplo um programa que leia um valor e que imprima a quantidade de cédulas necessárias para pagar esse mesmo valor. Para simplificar vamos trabalhar com valores inteiros e com cédulas de R$ 50, R$ 20, R$ 10, R$ 5 e R$ 1</p>

```python
valor = int(input("Digite o valor a pagar: "))
cedulas = 0
atual = 50
apagar = valor

while True:
    if atual <= apagar:
        apagar -= atual
        cedulas += 1

    else:
        print(f"{cedulas} cédula(s) de R${atual}")
        if apagar == 0:
            break
        if atual == 50:
            atual = 20
        elif atual == 20:
            atual == 10
        elif atual == 10:
            atual = 5
        elif atual == 5:
            atual = 1
        cedulas = 0
```

<h1 style="color:LightSeaGreen">Programa 5.5 - Repetição aninhadas</h1>
<p>Imprimir tabuada de multiplicação de 1 a 10</p>

```python
tabuada = 1
while tabuada <= 10:
    numero = 1
    while numero <= 10:
        print(f"{tabuada} x {numero} = {tabuada * numero}")
        numero += 1
    tabuada += 1
    
```

<p>Vejamos o mesmo programa, mas sem utilizar repetições aninhadas:</p>

```python
tabuada = 1
numero = 1
while tabuada <= 10:
    print(f"{tabuada} x {numero} = {tabuada * numero}")
    numero += 1
    if numero == 11:
        numero = 1
        tabuada += 1
```
